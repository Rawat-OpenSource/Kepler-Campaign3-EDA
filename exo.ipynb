```python
#Importing relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
test_data = pd.read_csv('/content/drive/MyDrive/exoTest.csv')
train_data = pd.read_csv('/content/drive/MyDrive/exoTrain.csv')
```

```python
#Basic information about the training data
print("Number of records:", len(train_data))
print("Number of attributes (columns):", len(train_data.columns))
```

```python
train_data.describe()
```

```python
#Correcting the values in the column "LABEL"
bin = {2: 1,1: 0}
train_data.LABEL = [bin[item] for item in train_data.LABEL]
test_data.LABEL = [bin[item] for item in test_data.LABEL]
```

```python
#Counting the number of examples for exoplanets vs not-exoplanets in the data
train_data['LABEL'].value_counts()
```

We observe that the dataset is highly imbalanced with 5050 records of not exoplantets as opposed to only 37 observations of exoplanets. Since the number of positive observations is very low as compared to the amound of negative observations, we would have to deal with the imbalance in the data before training our model.

```python
#PLotting a graph for the number of examples for exoplanets vs not exoplanets
plt.figure(figsize=(5,10))
colors = ["0", "1"]
sns.countplot(x = 'LABEL', data=train_data)
plt.title('Number of Observations \n (0: Not Exoplanet || 1: Exoplanet)', fontsize=12)
```

```python
# Checking for missing values
print("Total missing values in the dataset:")
print(train_data.isnull().sum().sum())
```

```python
#Comparisons between the transit curves of a recorded exoplanet and not exoplanet

fig, axes = plt.subplots(1, 2, sharey=True)

fig.set_figheight(5)
fig.set_figwidth(10)

axes[0].plot(range(train_data.shape[1] - 1), train_data.iloc[1, 1:])
axes[1].plot(range(train_data.shape[1] - 1), train_data.iloc[5085, 1:])

axes[0].set_title("Exoplanet example")
axes[1].set_title("Not exoplanet example")

axes[0].set_xlabel("Time")
axes[1].set_xlabel("Time")
axes[0].set_ylabel("Flux Intensity")

plt.tight_layout()
```

We can observe periodic drops in the flux intensity of the left plot which suggests the presence of an exoplanet, while the plot on the right mostly represents spikes in the flux intensity, largely caused by noise and suggest the absence of any exoplanets.

```python
#Comparisons between the transit curves of five recorded observations of an exoplanet

plt.figure(figsize=(15,10))
plt.title('Examples of distribution of Flux values for observed planets', fontsize=12)
plt.xlabel('Flux values')
plt.ylabel('Flux intensity')
plt.plot(train_data.iloc[0,])
plt.plot(train_data.iloc[1,])
plt.plot(train_data.iloc[2,])
plt.plot(train_data.iloc[3,])
plt.plot(train_data.iloc[4,])
plt.show()
```

The reccurent drops and spikes in the flux intensity recorded in the example of exoplanet observations show an interesting cyclic pattern.

```python
#Comparisons between the transit curves of five recorded observations of not an exoplanet

plt.figure(figsize=(15,10))
plt.title('Examples of distribution of Flux values for non-exoplanet observations', fontsize=12)
plt.xlabel('Flux values')
plt.ylabel('Flux intensity')
plt.plot(train_data.iloc[500,])
plt.plot(train_data.iloc[800,])
plt.plot(train_data.iloc[2300,])
plt.plot(train_data.iloc[300,])
plt.plot(train_data.iloc[700,])
plt.show()
```

The spikes and dips in the flux intensity recorded in the examples of non-exoplanet observations exhibit random patterns.
